```python
import pandas as pd
import numpy as np
from textblob import TextBlob
```

```python
twitter_data_df = pd.read_csv('twitter_data.csv', index_col=False)
```

```python
twitter_data_df = twitter_data_df.drop_duplicates('text')
twitter_data_df = twitter_data_df.reset_index()
```

```python
scores = []
for i in range(len(twitter_data_df)):
    scores.append(TextBlob(twitter_data_df['text'][i]).sentiment)
```

```python
sentiment = []
subjectivity = []
for score in scores:
    sentiment.append(score[0])
    subjectivity.append(score[1])
```

```python
twitter_data_df['sentiment_scores'] = sentiment
```

```python
twitter_data_df['subjectivity_scores'] = subjectivity
```

```python
twitter_data_df['date']
```

```python
twitter_data_df.to_csv('twitter_sentiment_data.csv', index=False)
```

```python
clean_dates = []
for i in range(len(twitter_data_df)):
    clean_dates.append(twitter_data_df['date'][i][4:10])
```

```python
twitter_data_df['clean_dates'] = clean_dates
```

```python
twitter_data_df = twitter_data_df[twitter_data_df.sentiment_scores != 0.0]
```

```python
twitter_data_df = twitter_data_df.reset_index()
```

```python
twitter_data_df['positive'] = twitter_data_df['sentiment_scores'] > 0
pd.DataFrame(twitter_data_df.groupby('clean_dates')['positive'].mean()).to_csv('twitter_sentiment_pos_ratio.csv')
```
